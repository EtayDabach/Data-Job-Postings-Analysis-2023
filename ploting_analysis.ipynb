# Plotting with python

```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from typing import Tuple
```

```python
# Query 1: Remote data analyst jobs
query_1 = pd.read_csv(r'queries_output_csv\1_remote_data_analyst_jobs_output.csv')
print(f'Table shape: {query_1.shape} as (rows, columns)')
query_1.head(5)
```

```python
# Query 2: Most demanded skills for query 1
query_2 = pd.read_csv(r'queries_output_csv\2_most_demanded_skills_for_query_1_output.csv')
print(f'Table shape: {query_2.shape} as (rows, columns)')
query_2.head(10)
```

```python
def barh_plot(vertical_axis, horizontal_axis, figsize:Tuple, title='', xlabel='', ylabel='', colordiv=False) -> Tuple[matplotlib.figure.Figure, matplotlib.axes._axes.Axes]:
    fig, ax = plt.subplots(figsize=figsize, facecolor='black')
    if colordiv:
        colorsgrad = horizontal_axis/np.max(horizontal_axis)
    else:
        colorsgrad = horizontal_axis
    ax.barh(vertical_axis, horizontal_axis, color=cm.Blues(colorsgrad), edgecolor='white')
    ax.invert_yaxis()
    ax.set_facecolor('black')
    ax.spines['left'].set_color('white')
    ax.spines['right'].set_color('white')
    ax.spines['top'].set_color('white')
    ax.spines['bottom'].set_color('white')
    ax.xaxis.label.set_color('white')
    ax.yaxis.label.set_color('white')
    ax.tick_params(axis='x', colors='white')
    ax.tick_params(axis='y', colors='white')
    ax.set_axisbelow(True)
    ax.grid()
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.set_title(title, color='white')
    plt.show()
    return fig, ax

def scatter_plot(x, y, figsize:Tuple, title='', xlabel='', ylabel='') -> Tuple[matplotlib.figure.Figure, matplotlib.axes._axes.Axes]:
    fig, ax = plt.subplots(figsize=figsize, facecolor='black')
    ax.scatter(x, y, s=100, linewidths=1, color='blue', edgecolors='white')
    ax.set_facecolor('black')
    ax.spines['left'].set_color('white')
    ax.spines['right'].set_color('white')
    ax.spines['top'].set_color('white')
    ax.spines['bottom'].set_color('white')
    ax.xaxis.label.set_color('white')
    ax.yaxis.label.set_color('white')
    ax.tick_params(axis='x', colors='white')
    ax.tick_params(axis='y', colors='white')
    ax.set_axisbelow(True)
    ax.grid()
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.set_title(title, color='white')
    plt.show()
    return fig, ax
```

```python
# Top 10 most demanded skills for remote data analyst
top_2 = 10
figsize_2 = (10, 6)
query_2_skills, query_2_demand_count = query_2['skills'][:top_2], query_2['demand_count'][:top_2]
title_2 = 'Top 10 most demanded skills for remote data analyst'
xlabel_2 = 'Demand Count'
fig_2, ax_2 = barh_plot(query_2_skills, query_2_demand_count, figsize_2, title_2, xlabel_2)
```

```python
# Top 10 most demanded skills average yearly salary for remote data analyst
query_2_avg = query_2['avg_salary'][:top_2]
title_2 = 'Top 10 most demanded skills average yearly salary for remote data analyst'
xlabel_2 = 'Average yearly salary [$]'
fig_2, ax_2 = barh_plot(query_2_skills, query_2_avg, figsize_2, title_2, xlabel_2, colordiv=True)
```
